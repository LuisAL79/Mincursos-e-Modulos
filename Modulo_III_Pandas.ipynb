```python
def pizza(colu,titulo):
  '''
  colu é a coluna de dados
  titulo é string - nome do gráfico
  '''
  dates = colu.value_counts().rename_axis('nomes').reset_index(name='counts')
  display(dates)

  v_values=dates.counts
  l_labels=dates.nomes

  fig, ax = plt.subplots(figsize=(8,8),facecolor='lightblue')

  ax.pie(v_values,labels=l_labels,autopct='%1.1f%%',\
        shadow=False, startangle=-90,textprops=dict(color="k",\
                                                    fontsize=16,weight="bold"))
  ax.axis('equal')
  ax.set_title(titulo,weight='bold',size=18,color='red')

  plt.show()
```

```python
pizza(dat_df['GRAU'],'Grau oferecido')
```
