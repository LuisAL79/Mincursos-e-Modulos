```python
xc = np.linspace(-3*np.pi,3*np.pi,70)
yc = np.cos(xc)
ys = np.sin(xc)

def funsc(xc):
  return np.sin(xc)+np.cos(xc)

plt.plot(xc,yc,'or',label='cos')
plt.plot(xc,funsc(xc),'-y',label='sum')

plt.xlabel(r'$\theta$')
plt.ylabel('function')
plt.legend()
plt.show()
```

## **Tamanho e multiplots**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xc = np.linspace(-np.pi,np.pi,70)
yc = np.cos(xc)
ys = np.sin(xc)

fig1=plt.figure()
plt.plot(xc,ys)

fig2=plt.figure()
plt.plot(xc,yc)
plt.grid('True')

fig3=plt.figure()
plt.plot(ys,yc)
plt.axis('equal')
plt.grid('True')

plt.show()
```

```python
fig = plt.figure()

ax1=plt.subplot(221)
plt.plot(xc,yc)
plt.title('figura 1')


ax2=plt.subplot(222)
plt.plot(xc,ys)
plt.title('figura 2')

ax3=plt.subplot(224)
plt.plot(yc,ys)
plt.axis('equal')

plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,5), sharey=True)

ax[0].plot(xc,ys)
ax[0].set_ylabel('y')

ax[1].plot(xc,1*yc,'-r')

plt.show()
```

```python
def f(t):
    s1 = np.cos(2*np.pi*t)
    e1 = np.exp(-t)
    return s1 * e1

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
t3 = np.arange(0.0, 2.0, 0.01)

fig, axs = plt.subplots(1,2, constrained_layout=True,
                        figsize=(10,5) )

axs[0].plot(t1, f(t1), 'o', t2, f(t2), '-')
axs[0].set_title('subplot 1')
axs[0].set_xlabel('distance (m)')
axs[0].set_ylabel('Damped oscillation')

axs[1].plot(t3, np.cos(2*np.pi*t3), '--')
axs[1].set_xlabel('time (s)')
axs[1].set_title('subplot 2')
axs[1].set_ylabel('Undamped')

fig.suptitle('Figura com t√≠tulo superior', fontsize=16)

plt.savefig('figura-minha.png')
plt.show()
```

## **Param√™tricas**

$$x=r\cos(t)$$
$$y=r\sin(t)$$

```python
def trac(t):
  return np.array([np.cos(t), np.sin(t)])

t=np.linspace(0,3*np.pi/2,200)

x,y=trac(t)

plt.plot(x,y)
plt.axis('equal')
plt.show()
```

```python
def traje(t):
  return np.array([t*np.cos(t),np.sin(2*t)])

t=np.linspace(0,20,200)

x,y=traje(t)

plt.plot(x,y)
plt.show()
```

## **Campo Vetorial**

$$\vec{F}(x,y)=-y\hat{i}+x\hat{j}$$

----

image.png

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def campo(x,y):
  return np.array([ -y, x ])

x=np.linspace(-2,2,20)
y=np.linspace(-2,2,20)

xmesh,ymesh=np.meshgrid(x,y)     # mesh: malha  | meshgrid: grade de malha
umesh,vmesh=campo(xmesh,ymesh)

fig, ax=plt.subplots(1,1 , figsize=(6,5))
ax.set_title('Arrows with plot width')

Q=ax.quiver(xmesh,ymesh,umesh,vmesh,color='k',
            units='width',angles='xy')
qk = ax.quiverkey(Q, 0.75, 0.91, 2, r'$2 \frac{m}{s}$',
                  labelpos='E', coordinates='figure')

ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.set_aspect('equal')

plt.show()
```

## **Curvas 3d**

```python
from mpl_toolkits.mplot3d import axes3d

def tra3d(t):
  return np.array([ t*np.sin(t), t*np.cos(t), t ])

t=np.linspace(-15,15,200)
x,y,z=tra3d(t)

fig=plt.figure(figsize=(9,6))
ax=plt.axes(projection='3d')
ax.plot(x,y,z,label='espiral')
ax.set_zlabel('f(t)')
ax.legend()

plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d

def fun3d(x,y):
  return np.sin(np.sqrt(x*x+y*y))

x = np.linspace(-4,4,25)
y = np.linspace(-5,5,25)

xmesh,ymesh=np.meshgrid(x,y)
zmesh=fun3d(xmesh,ymesh)

fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')
plt.rcParams['legend.fontsize'] = 14

surf=ax.plot_surface(xmesh,ymesh,
                     zmesh,label=r'$\sin(\sqrt{x^2+y^2})$',
                     color='red')
surf._facecolors2d = surf._facecolor3d
surf._edgecolors2d = surf._edgecolor3d

ax.legend()
plt.show()
```
